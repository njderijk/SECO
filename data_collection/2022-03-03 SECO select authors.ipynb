```python
authorDf = authorDf.drop_duplicates(subset=['authorID', 'methodHash', 'projectVersion'])
authorDf
```

```python
authorDf = authorDf.drop_duplicates(subset=['authorID', 'methodHash', 'projectID'], keep='last')
authorDf
```

```python
authorDf.to_csv('cleanedAuthorMethodV2.csv')
```

```python
uniqueAuthor = authorDf.groupby('authorID')
```

```python
len(uniqueAuthor)
```

```python
authProjDictList = []

for name, group in uniqueAuthor:
    authProjDict = {}
    uniqueProjects = group['projectID'].nunique()
    authProjDict['author'] = name
    authProjDict['noOfProj'] = uniqueProjects
    
    authProjDictList.append(authProjDict)
```

```python
len(authProjDictList)
```

```python
#DEZE HOEFT DUS NIET MEER
multiProjAuth = []

for authorProj in authProjDictList:
    if authorProj['noOfProj'] > 1:
        multiProjAuth.append(authorProj)
```

```python
authProjDictList[:10]
```

```python
i = 0
begin = time.time()
for authorDict in authProjDictList:
    author = authorDict['author']
    noOfMethods = authorDf['authorID'].value_counts()[author]
    authorDict['noOfMethods'] = noOfMethods
    if i % 100 == 0:
        if i > 0:
            dur = time.time() - begin
            timeToGo = ((len(authProjDictList) - i) * dur) / i
            hoursToGo = timeToGo / 60 / 60
            print(i, ' authors gedaan')
            print('Nog ongeveer', hoursToGo, ' uur te gaan')
    i += 1
```

```python
summaryDf = pd.DataFrame(authProjDictList)
```
