```python
import socket
import pandas as pd
import time
import datetime
```

```python
host = '95.217.155.82'
port = 8003
```

```python
def make_request(host, port, request):
    s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    s.connect((host, port))

    byt = request.encode()
    s.send(byt)
    
    answer = request_loop(s)

    s.close()
    
    return answer
```

```python
def request_loop(s, timeout=2):
    s.setblocking(0)
    total_data = []
    data = ''
    begin = time.time()

    while 1:
        if total_data and time.time()-begin > timeout:
            break
        elif time.time()-begin > timeout*2:
            break
        
        try:
            data = s.recv(4096)
            if data:
                data = data.decode('utf-8')
                total_data.append(data)
                begin = time.time()
            else:
                time.sleep(0.1)
        except:
            pass

    return ''.join(total_data)
```

```python
def clean_ans(answer):
    answer = answer.splitlines()
    answer.pop(0)
    
    return answer
```

```python
def create_list(answer):
    answerList = []
    if answer[0] == 'No results found.':
        return 'no'
    else:
        for i in answer:
            el = i.split('?')
            answerList.append(el)

    return answerList
```
